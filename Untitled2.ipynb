```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

# ---------------------------------------------------------
# 1. Generate Independent Synthetic Data
# ---------------------------------------------------------
np.random.seed(42)
n_students = 600

# Creating realistic distributions for student metrics
study_time = np.random.normal(5, 2, n_students).clip(0, 10)  # Hours per week
attendance = np.random.normal(85, 10, n_students).clip(30, 100) # Percentage
past_grades = np.random.normal(70, 15, n_students).clip(0, 100) # Out of 100

df = pd.DataFrame({
    'Study_Time': study_time,
    'Attendance': attendance,
    'Past_Grades': past_grades
})

# Define Pass/Fail based on a weighted combination of features
performance_score = (df['Study_Time']*3 + df['Attendance']*0.4 + df['Past_Grades']*0.3)
# Top 60% pass, bottom 40% fail
threshold = np.percentile(performance_score, 40)
df['Pass_Exam'] = (performance_score >= threshold).astype(int)

# ---------------------------------------------------------
# 2. Exploratory Data Analysis (EDA)
# ---------------------------------------------------------
print("--- Exploratory Data Analysis ---")
print(df.describe().round(2))

plt.figure(figsize=(10, 6))
sns.pairplot(df, hue='Pass_Exam', palette='Set1', diag_kind='kde')
plt.suptitle('EDA: Feature Relationships by Exam Outcome', y=1.02)
plt.show()

# ---------------------------------------------------------
# 3. Unsupervised Clustering (K-Means)
# ---------------------------------------------------------
# Standardize features for clustering
features = ['Study_Time', 'Attendance', 'Past_Grades']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])

# Apply K-Means
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Label Clusters based on centroid values (Higher means consistent)
centroids = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=features)
cluster_means = centroids.mean(axis=1)

consistent_cluster_id = cluster_means.idxmax()
df['Profile'] = df['Cluster'].apply(
    lambda x: 'Consistent Learner' if x == consistent_cluster_id else 'Irregular Learner'
)

print("\n--- Clustering Results ---")
print(df['Profile'].value_counts())
print("\nCluster Centroids (Original Scale):")
print(centroids.round(2))

# ---------------------------------------------------------
# 4. Supervised Models & Cross-Validation
# ---------------------------------------------------------
X = X_scaled
y = df['Pass_Exam']

# Define models
models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100)
}

print("\n--- Supervised Model Cross-Validation ---")
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    # 5-fold cross-validation for Accuracy, Precision, and Recall
    acc_scores = cross_val_score(model, X, y, cv=cv, scoring='accuracy')
    f1_scores = cross_val_score(model, X, y, cv=cv, scoring='f1')

    print(f"\n{name}:")
    print(f"Mean Accuracy: {acc_scores.mean():.4f} (+/- {acc_scores.std() * 2:.4f})")
    print(f"Mean F1-Score: {f1_scores.mean():.4f}")
```
